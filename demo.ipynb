```python
source = pd.read_csv("input/worldcities.csv")
source.country = source.country.replace("Czechia", "Czech Republic")
source['in_europe'] = source.country.apply(lambda x: x in european_countries)
source = source.dropna(subset=('population')).reset_index(drop=True)
source
```

```python
choosen_cities = {}

expat_countries = {
    'Canada': 8,
    'United Arab Emirates': 5,
    'Israel': 3,
    'Saudi Arabia': 5, 
    'Qatar': 2,
    'United States': 50,
    'Australia': 5,
    'Japan': 9,
    'Thailand': 5,
    'Philippines': 2,
    'Turkey': 5,
    'Mexico': 5,
    'Korea, South': 15,
    'Vietnam': 10,
    'Argentina': 2,
    'Taiwan': 1,
    'Malaysia': 1,
    'Hong Kong': 1,
}
large_countries = {
    'China': 10,
    'India': 8,
    'Brazil': 5,
    'Pakistan': 2,
    'Iran': 2,
    'Indonesia': 3,
    'Egypt': 3,
    'Colombia': 2,
    'South Africa': 2,
}

slavic_countries = {
    "Russia": 20,
    'Ukraine': 10,
    "Belarus": 3,
}

central_asia_countries = {
    "Kyrgyzstan": 3, 
    "Uzbekistan": 16,
    "Armenia": 4, 
    "Georgia": 4, 
    "Azerbaijan": 4, 
    "Turkmenistan": 3, 
}
# add kazakhstan
kz_citie_df = source[source.country == 'Kazakhstan'].head(30)
choosen_cities['kz_cities'] = kz_citie_df.id.tolist()

# add europe
european_cities_df = source[(source.in_europe)].head(100)
choosen_cities['european_cities'] = european_cities_df.id.tolist()

expat_cities = []
for country, n_cities in expat_countries.items():
    sample = source.loc[source.country == country, 'id'].head(n_cities).tolist()
    expat_cities.extend(sample)
choosen_cities['expat_cities'] = expat_cities

central_asia_cities = []
for country, n_cities in central_asia_countries.items():
    sample = source.loc[source.country == country, 'id'].head(n_cities).tolist()
    central_asia_cities.extend(sample)
choosen_cities['central_asia_cities'] = central_asia_cities

slavic_cities = []
for country, n_cities in slavic_countries.items():
    sample = source.loc[source.country == country, 'id'].head(n_cities).tolist()
    slavic_cities.extend(sample)
choosen_cities['slavic_cities'] = slavic_cities

large_countries_cities = []
for country, n_cities in large_countries.items():
    sample = source.loc[source.country == country, 'id'].head(n_cities).tolist()
    large_countries_cities.extend(sample)
choosen_cities['large_countries_cities'] = large_countries_cities

choosen_cities.keys()
```

```python
from IPython.display import display
priority_group = []
non_priority_group = []
priority_thresholds = {
    'kz_cities': 200,
    'european_cities': 1000,
    'expat_cities': 1500,
    'central_asia_cities': 350,
    'slavic_cities': 1100,
    'large_countries_cities': 20000,
}
for group in choosen_cities.keys():
    cities_ids = choosen_cities[group]
    group_df = source[source.id.isin(cities_ids)].sort_values('population', ascending=False)
    print(cities_ids)
    
    threshold = priority_thresholds.get(group, None)
    if threshold is None:
        continue
    priority_group_df = group_df[group_df.population >= threshold * 1000]
    non_priority_group_df = group_df[group_df.population < threshold * 1000]
    # display(priority_group_df)
    # display(non_priority_group_df)

    priority_group.append(priority_group_df)
    non_priority_group.append(non_priority_group_df)

index_df = pd.concat(priority_group + non_priority_group).reset_index(drop=True)
index_df
```

```python
other_countries = source[~source.country.isin(index_df.country.unique())].country.unique()
other_countries_df = source[source.country.isin(other_countries)]
top_other_countries = other_countries_df.groupby('country').population.sum().sort_values(ascending=False).head(30).index.tolist()
other_cities = []
for country in top_other_countries:
    print(country)
    top_city = other_countries_df[other_countries_df.country == country].iloc[0:1]
    other_cities.append(top_city)
other_countries_index = pd.concat(other_cities).sort_values('population', ascending=False)
index_df = pd.concat((index_df, other_countries_index)).reset_index(drop=True)
index_df
```
